# Puzzle

Find all 3-digit numbers so that the number is equal to the sum of the factorial of its digits.

```python
from math import factorial as fact
```

### Solution 1

```python
def find():
    numbers = range(100, 999)
    required_num = []
    for n in numbers:
        digits = [int(d) for d in str(n)]
        factorial_sum = fact(digits[0]) + fact(digits[1]) + fact(digits[2])
        if factorial_sum == n:
            required_num.append(n)
    return required_num
```

```python
find()
```
